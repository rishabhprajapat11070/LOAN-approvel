```python
import pandas as pd
data = pd.read_csv("Loan_approval_data_2025.csv")
data.info()
data.head(10)
```

```python
newdata = data[["credit_score","defaults_on_file","delinquencies_last_2yrs","derogatory_marks","annual_income","current_debt","debt_to_income_ratio","loan_amount","interest_rate","credit_history_years","loan_status"]]
newdata.head()
pd.isnull(newdata).sum()
newdata.drop(columns=["defaults_on_file"],inplace=True)
newdata
```

```python
from sklearn.model_selection import train_test_split

Inputx = newdata[["credit_score","delinquencies_last_2yrs","derogatory_marks","annual_income","current_debt","debt_to_income_ratio","loan_amount","interest_rate","credit_history_years"]]
outputy = newdata["loan_status"]

x_train,x_test,y_train,y_test = train_test_split(Inputx,outputy,test_size=0.2,random_state=42)

x_test.shape,x_train.shape

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

scaler = StandardScaler()
scaledXtrain= scaler.fit_transform(x_train)
scaledXtest = scaler.transform(x_test)

mmodel = LogisticRegression()
mmodel.fit(scaledXtrain,y_train)

Input = [[627,1,0,43087,6550,0.384,53300,14.10,3.5]]
inscaler = scaler.transform(Input)
mmodel.score(scaledXtest,y_test)
pre = mmodel.predict(inscaler)

if pre == 1:
    print("===============")
    print("===============")
    print("appreoved")
    print("===============")
    print("===============")
else:
    print("not approved")



```
